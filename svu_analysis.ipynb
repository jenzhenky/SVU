## SVU Cast Dataframe Manipulation

#### Code to generate a dataframe with boolean True/False values for every actor/character combination in each episode of SVU

```python
import pandas as pd
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)

import matplotlib as plt
%matplotlib inline

import re

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
```

```python
df_svu_raw = pd.read_csv(r'C:\Users\Jenny\svu_cleaned.csv')
df_svu_raw = df_svu_raw.drop('Unnamed: 0', axis=1)
df_svu_raw['season'] = df_svu_raw['ep_id'].apply(lambda x: int(x[:2]))
df_svu_raw.head(5)
```

```python
# Format the actor and character lists, then create a list that combines the two as "actor_as_character"
df_svu_raw['actor'] = df_svu_raw['actor'].apply(lambda x: x.strip('[]').split(','))
df_svu_raw['actor'] = df_svu_raw['actor'].apply(lambda x: [re.sub(r"^'|'$", "", item) for item in x])
df_svu_raw['actor'] = df_svu_raw['actor'].apply(lambda x: [re.sub(r"\s'|'\s", " ", item) for item in x])
df_svu_raw['character'] = df_svu_raw['character'].apply(lambda x: x.strip('[]').split(','))
df_svu_raw['character'] = df_svu_raw['character'].apply(lambda x: [re.sub(r"^'|'$", "", item) for item in x])
df_svu_raw['character'] = df_svu_raw['character'].apply(lambda x: [re.sub(r"\s'|'\s", " ", item) for item in x])
df_svu_raw['actor_as_role'] = df_svu_raw.apply(lambda x: [(actor + '_as_' + role) for actor, role in zip(x['actor'], x['character'])], axis=1)
df_svu_raw['actor_as_role'] = df_svu_raw['actor_as_role'].apply(lambda x: [str(item.replace('"','').replace("_as_ ", "_as_")) for item in x])
df_svu_raw['actor_as_role'] = df_svu_raw['actor_as_role'].apply(lambda x: [item.lstrip() for item in x])
df_svu_raw['actor_as_role'].head(1)
```

```python
# Create a list of all unique actor-character combinations
list_actor_roles = list(set([item for sublist in list(df_svu_raw['actor_as_role']) for item in sublist]))
list_actor_roles = list(set([x.lstrip() for x in list_actor_roles]))
list_actor_roles.remove('Steven Zirnkilton_as_Opening Announcer')
list_actor_roles.sort()
list_actor_roles
```

```python
# Create a dictionary of {episode_id: actor_as_role}
dict_episodes = pd.Series(df_svu_raw['actor_as_role'].values, index=df_svu_raw['ep_id']).to_dict()
dict_episodes
```

```python
# Create a dataframe df_svu_eps with list_actor_roles
df_svu_eps = pd.DataFrame(list_actor_roles, columns=['actor_as_role'])
df_svu_eps.head()
```

```python
# Split [actor_as_role] into [actor] and [character]
df_svu_eps['actor_raw'] = df_svu_eps['actor_as_role'].apply(lambda x: x.split('_as_'))
df_svu_eps['actor'] = df_svu_eps['actor_raw'].apply(lambda x: x[0])
df_svu_eps['role'] = df_svu_eps['actor_raw'].apply(lambda x: x[1])
del df_svu_eps['actor_raw']
```

```python
# Loop through the dictionary to check if each actor-character pair is in each episode. True is 1 and False is 0.
for key, val in dict_episodes.items():
    df_svu_eps[key] = df_svu_eps['actor_as_role'].apply(lambda x: int(x in val))
```

```python
# Check that df_svu_eps now contains episode data for each actor-character pair 
df_svu_eps[df_svu_eps['actor_as_role'] == "Christopher Meloni_as_Elliot Stabler"]
```

```python
df_svu_eps['ep_id'] = df_svu_eps.drop(['actor_as_role','actor','role'], axis=1).idxmax(axis=1)
```

### Who's been in the most episodes of SVU?

```python
df_svu_eps['total_eps'] = df_svu_eps.sum(numeric_only=True, axis=1)
```

```python
# Sort by total_eps to find total episodes with each actor-role pair
df_svu_eps[['actor_as_role','total_eps']].sort_values('total_eps', ascending=False)
```

```python
# Sort by total_eps to find total episodes with each actor
df_actor = df_svu_eps.groupby('actor')['total_eps'].sum().sort_values(ascending=False)
df_actor
```

```python
# Bin df_actor_role to see the distribution of how often actor-role pairs appear on the show
bins = [0, 5, 10, 20, 100, 500]
df_actor.groupby(pd.cut(df_actor, bins=bins)).size()
```
